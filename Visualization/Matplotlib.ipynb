```python
import matplotlib as plt
import numpy as np
import pandas as pd
```

```python
df2=pd.read_csv("D:\\Data Analyst\\ML\\ML Tutorial\\dataset\\HR_Analytics.csv")
```

```python
df2
```

```python
income_by_gender = df2.groupby("Gender")["MonthlyIncome"].sum()

plt.pyplot.pie(income_by_gender, 
        labels=income_by_gender.index, 
        autopct='%1.1f%%')

plt.pyplot.title("Monthly Income Distribution by Gender")
plt.pyplot.show()
```

```python
plt.pyplot.bar(df2["JobRole"], df2["MonthlyIncome"])
plt.pyplot.xlabel("Job Role")
plt.pyplot.ylabel("Monthly Income")
plt.pyplot.title("Monthly Income by Job Role")
plt.pyplot.xticks(rotation=45)
plt.pyplot.show()
```

```python
plt.pyplot.scatter(df2["Age"], df2["MonthlyIncome"])
plt.pyplot.xlabel("Age")
plt.pyplot.ylabel("Monthly Income")
plt.pyplot.title("Monthly Income vs Age")
plt.pyplot.show()
```
