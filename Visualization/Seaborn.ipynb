```python
import seaborn as sns
import matplotlib as plt
import pandas as pd
import numpy as np
```

```python
df=sns.load_dataset("titanic")
df.head()
```

```python
df.isnull().sum()
```

```python
df["age"].fillna(df["age"].mean(), inplace=True)
df["embarked"].fillna(df["embarked"].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
sns.set_style("darkgrid")

sns.scatterplot(x="age", y="fare", data=df,hue="sex")
plt.pyplot.title("Age vs Price")
plt.pyplot.xlabel("Age")
plt.pyplot.ylabel("price")
plt.pyplot.show()
```

```python
df["Survived"]=df["survived"].map({0:"No",1:"Yes"})
sns.countplot(x="Survived", data=df,hue="sex")
plt.pyplot.title("Survival Count")
plt.pyplot.xlabel("Survived")
plt.pyplot.ylabel("Count")
plt.pyplot.show()
```

```python
sns.swarmplot(x="class", y="age", data=df,hue="sex")
```

```python


#showing the normal distribution of age for each gender
sns.histplot(x="age", data=df, bins=30, kde=True,hue="sex")
```

```python
sns.pairplot(df,hue="sex")
```

```python
sns.lineplot(x="age", y="Survived", data=df,hue="sex"   )
```
