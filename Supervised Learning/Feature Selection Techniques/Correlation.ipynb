```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
```

```python
data=load_breast_cancer()
```

```python
x=pd.DataFrame(data.data,columns=data.feature_names)
y=pd.DataFrame(data.target,columns=['target'])
```

```python
corr_matrix=x.corr()
print(corr_matrix)
```

```python
upper_matrix=corr_matrix.where(np.triu(np.ones(corr_matrix.shape),k=1).astype(bool))
drop_features=[column for column in upper_matrix.columns if any(upper_matrix[column].abs()>0.9)]
print("Drop features:",drop_features)

selected_features=x.drop(drop_features,axis=1)
print("Selected features:",selected_features.columns.tolist())
```
